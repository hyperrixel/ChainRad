# Import section

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import auc, confusion_matrix, precision_recall_curve, roc_curve
```

# Constants and variables

```python
diseasesonly = ["infiltration", "effusion", "atelectasis", "consolidation", "mass",
                "nodule", "edema", "pneumothorax", "cardiomegaly", "emphysema", "pleural_thickening",
                "pneumonia", "fibrosis", "hernia"]
th = 0.5
precision = None
recall = None
thresholds = None
```

# Functions

```python
def calculate_f1score(precision, recall):
    
    if (precision + recall) == 0:
        return 0.0
    else:
        f1score = 2 * (precision * recall) / (precision + recall)
        return f1score

    
def f1(threshold):
    for i,t in enumerate(thresholds):
        if t > threshold:
            f1score = calculate_f1score(precision[i], recall[i])
            if not math.isnan(f1score):
                return f1score
            else: 
                return 0
    return 0


def plot_auc(t_y, p_y):

    false_positive_rate, true_positive_rate, thresholds = roc_curve(t_y, p_y, pos_label=1)
    plt.plot(false_positive_rate, true_positive_rate, color='orange', lw=2)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')    
    plt.show()


def plot_history(history):
    
    f = plt.figure()
    f.set_figwidth(20)
    f.add_subplot(1, 2, 1)
    plt.plot(history.history['val_loss'], label='Validation loss')
    plt.plot(history.history['loss'], label='Training loss')
    plt.legend()

    f.add_subplot(1, 2, 2)
    plt.plot(history.history['val_binary_accuracy'], label='Validation accuracy')
    plt.plot(history.history['binary_accuracy'], label='Training accuracy')
    plt.legend()
    plt.show()


def plot_precision(t_y, p_y):
       
    precision, recall, _ = precision_recall_curve(t_y, p_y, pos_label=1)
    plt.plot(recall, precision, color='orange', lw=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.show()

    
def plot_thresholds(t_y, p_y):
    
    global precision
    global recall
    global thresholds
    
    precision, recall, thresholds = precision_recall_curve(t_y, p_y, pos_label=1)
    plt.plot(thresholds, precision[:-1], color='orange', lw=2, label='Precision')
    plt.plot(thresholds, recall[:-1], color='navy', lw=2, label='Recall')
    plt.legend()
    plt.xlabel('Thresholds')
    plt.ylabel('Values')
    plt.title('Precision/Recall by Thresholds')
    plt.show()
```

# Plot scores and get optimal threshold

```python
for disease in diseasesonly:
    print(disease)
    temp_df = pd.read_csv('./logs/final_{}.csv'.format(disease), sep='\t')
    valY = temp_df['target']
    pred_Y_temp = temp_df['prediction']
    pred_Y = []
    pred_Y = pred_Y_temp
    
    plot_auc(valY, pred_Y)
    plot_precision(valY, pred_Y)
    plot_thresholds(valY, pred_Y)
    
    f1scores = [f1(t) for t in thresholds]
    plt.plot(thresholds, f1scores, color='darkgreen')
    plt.title('F1 Scores by Threshold')
    plt.xlabel('threshold')
    plt.ylabel('F1')
    plt.show()
    
    i = np.argmax(np.array(f1scores, dtype=np.float32))
    f1val = f1scores[i]
    
    print()
    print('F1 score max: {:6.4f}'.format(f1val))
    print('   Threshold: {:6.4f}'.format(thresholds[i]))
    print('   Precision: {:6.4f}'.format(precision[i]))
    print('      Recall: {:6.4f}'.format(recall[i]))
    
    plt.plot(thresholds, recall[:-1], color='navy', lw=2, label='Recall')
    plt.plot(thresholds, f1scores, label='F1', color='darkgreen')
    plt.title('F1 Scores and Threshold')
    plt.xlabel('Threshold')
    plt.legend()
    plt.show()

    
    threshold = thresholds[i]

    alt_pred_y = []

    for i in pred_Y_temp:
        if i >= threshold:
            alt_pred_y.append(1)
        else:
            alt_pred_y.append(0)


    target_labels = [0, 1]

    true_negative, false_positive, false_negative, true_positive = confusion_matrix(valY, alt_pred_y, target_labels).ravel()

    labels = ['negative', 'positive']
    cm = confusion_matrix(valY,  alt_pred_y, target_labels)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(cm)
    plt.title('Confusion matrix of the classifier\n').set_fontsize(20)
    fig.colorbar(cax)
    for i in range(2):
        for j in range(2):
            plt.text(j, i, "{}".format(cm[i, j]),
                         horizontalalignment="center",
                         color="white").set_fontsize(20)
    ax.set_xticklabels([''] + labels)
    ax.set_yticklabels([''] + labels)        
    plt.xlabel('Predicted').set_fontsize(14)
    plt.ylabel('True').set_fontsize(14)
    plt.show()

    print('{:>18}True Negative: {}'.format('', true_negative))
    print('{:>18}True Positive: {}'.format('', true_positive))
    print('{:>17}False Negative: {}'.format('', false_negative))
    print('{:>17}False Positive: {}'.format('', false_positive))
```
