# Import section

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Constants and variables

```python
diseasesonly = ["infiltration", "effusion", "atelectasis", "consolidation", "mass",
                "nodule", "edema", "pneumothorax", "cardiomegaly", "emphysema", "pleural_thickening",
                "pneumonia", "fibrosis", "hernia"]
```

# Functions

```python
def plot_history(history):
    
    train_loss = list(history['train_loss'])
    test_loss = list(history['test_loss'])
    train_accuracy = list(history['train_accuracy'])
    test_accuracy = list(history['test_accuracy'])
    
    f = plt.figure()
    f.set_figwidth(20)
    f.add_subplot(1, 2, 1)
    plt.plot(test_loss, label='Testing loss')
    plt.plot(train_loss, label='Training loss')
    plt.legend()

    f.add_subplot(1, 2, 2)
    plt.plot(test_accuracy, label='Testing accuracy')
    plt.plot(train_accuracy, label='Training accuracy')
    plt.legend()
    plt.show()    
```

# Plot training and testing history

```python
for disease in diseasesonly:
    temp_df = pd.read_csv('./logs/{}.csv'.format(disease), sep='\t')
    print(disease)
    plot_history(temp_df)
```
